```python
"Example Sudoku"
Initial = ([0,0,0,2,6,0,7,0,1],
           [6,8,0,0,7,0,0,9,0],
           [1,9,0,0,0,4,5,0,0],
           [8,2,0,1,0,0,0,4,0],
           [0,0,4,6,0,2,9,0,0],
           [0,5,0,0,0,3,0,2,8],
           [0,0,9,3,0,0,0,7,4],
           [0,4,0,0,5,0,0,3,6],
           [7,0,3,0,1,8,0,0,0])
Sudoku = Initial
```

```python
"Checks vertical, horizontal and blocks for duplicates"
def Check(var):
    Sudoku,i,j = var
    row_list = [0]*9
    pos1 = 0
    pos2 = 0
    pos3 = 0
    
    "Check Columns"
    for x in range(9):
        if pos1 == j:          #check that we don't test the position that we are changing
            h = 0
        else:
            if Sudoku[i][x] == Sudoku[i][j]:
                condition1 = False
                break
            else:
                condition1 = True
        pos1 = pos1 + 1
        
    "Check row"            
    for z in range(9):
        row_list[z] = Sudoku[z][j]
        
    for y in range(9):
        if pos2 == i:
            h = 0
        else:
            if row_list[y] == Sudoku[i][j]:
                condition2 = False
                break
            else:
                condition2 = True
        pos2 = pos2 + 1
    "Square Check"
    matrix_list, ans = small_sudoku([Sudoku,i,j])
    for v in range(9):
        if v == ans:
            h = 0
        else:
            if matrix_list[v] == Sudoku[i][j]:
                condition3 = False
                break
            else:
                condition3 = True
    
    
    return condition1, condition2, condition3
```

```python
"Main function that solves the Sudoku"
def Solver(Sudoku):
    for i in range(9):            #Row
        for j in range(9):        #Column
            if Initial[i][j] == 0:
                for k in range(1,10):
                    Sudoku[i][j] = k    #Insert 
                    Answer = Check([Sudoku,i,j])
                    if Answer[0] == True and Answer[1] == True and Answer[2] == True:
                        break
                        
    return Sudoku
```

```python
"Sets up the 3x3 blocks inside the sudoku"
def small_sudoku(var):
    Sudoku, i, j = var
    matrix_list = [0]*9
    count = 0
    if i<3 and j<3:
        for x in range(0,3):
            for y in range(0,3):
                matrix_list[count] = Sudoku[x][y]
                
                if x==i and y == j:
                    ans = count
                count = count + 1
                
    elif i<3 and 3<=j<6:
        for x in range(0,3):
            for y in range(3,6):
                matrix_list[count] = Sudoku[x][y]
                
                
                if x==i and y == j:
                    ans = count
                count = count + 1
                
    elif i<3 and 6<=j<9:
        for x in range(0,3):
            for y in range(6,9):
                matrix_list[count] = Sudoku[x][y]
                
                
                if x==i and y == j:
                    ans = count
                count = count + 1
                
    elif 3<=i<6 and j<3:
        for x in range(3,6):
            for y in range(0,3):
                matrix_list[count] = Sudoku[x][y]
                                
                
                if x==i and y == j:
                    ans = count
                count = count + 1
                
    elif 3<=i<6 and 3<=j<6:
        for x in range(3,6):
            for y in range(3,6):
                matrix_list[count] = Sudoku[x][y]
                 
                
                if x==i and y == j:
                    ans = count
                count = count + 1
                
    elif 3<=i<6 and 3<=j<9:
        for x in range(3,6):
            for y in range(6,9):
                matrix_list[count] = Sudoku[x][y]
                
                
                if x==i and y == j:
                    ans = count
                count = count + 1
                
    elif 6<=i<9 and j<3:
        for x in range(6,9):
            for y in range(0,3):
                matrix_list[count] = Sudoku[x][y]
                                
                if x==i and y == j:
                    ans = count
                count = count + 1
                    
    elif 6<=i<9 and 3<=j<6:
        for x in range(6,9):
            for y in range(3,6):
                matrix_list[count] = Sudoku[x][y]
                                
                
                if x==i and y == j:
                    ans = count
                count = count + 1
                    
    elif 6<=i<9 and 6<=j<9:
        for x in range(6,9):
            for y in range(6,9):
                matrix_list[count] = Sudoku[x][y]
                
                                
                
                if x==i and y == j:
                    ans = count
                count = count + 1
                    
    return matrix_list, ans
```
